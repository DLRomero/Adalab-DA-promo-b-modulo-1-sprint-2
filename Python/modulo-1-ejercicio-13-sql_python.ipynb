```python
# Extraed el primer pedido donde se ha gastado más. Devolved el id del pedido y la cantidad gastada.
#fetchone
import mysql.connector

mydb = mysql.connector.connect(
                  user='root', password='AlumnaAdalab',
                  host='127.0.0.1',
                  database='northwind')

mycursor = mydb.cursor()

mycursor.execute("SELECT order_id, SUM((unit_price*quantity)/(1+discount)) FROM order_details GROUP BY order_id ORDER BY SUM((unit_price*quantity)/(1+discount)) DESC LIMIT 1;")

for x in mycursor:
  print(x)
```
